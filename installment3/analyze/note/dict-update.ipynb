```python
import pandas as pd
import yaml
```

```python
df = pd.read_csv('../input/ARTS_Passenger-redacted.csv.gz', sep='|')
```

```python
df.info()
```

```python
df.head()
```

```python
with open('../input/airport_dict.yaml', 'r') as yamlfile:
    air_dict = yaml.load(yamlfile)
```

```python
known_airports = set(air_dict.keys())
```

```python
air_dict
```

```python
name_dict = {v['AirportName']: k for k, v in air_dict.items()}
```

```python
df['PULOC'] = df['air_AirportName'].replace(name_dict)
df['DropLoc'] = df['air2_AirportName'].replace(name_dict)
```

```python
i3_puloc = set(df['PULOC'].unique())
```

```python
i3_puloc_new = i3_puloc.difference(known_airports)
```

```python
i3_puloc_new
```

```python
i3_droploc = set(df['DropLoc'].unique())
```

```python
i3_droploc_new = i3_droploc.difference(known_airports)
```

```python
i3_droploc_new.union(i3_puloc_new)
```

```python
new_airports = pd.read_csv('../hand/i3_new_airports.csv')
```

```python
new_airports = new_airports.set_index('ICAOCode')
```

```python
new_airports = new_airports.to_dict(orient='index')
```

```python
pulocs = df.groupby('MissionID')['PULOC'].unique().reset_index()
droplocs = df.groupby('MissionID')['DropLoc'].unique().reset_index()
```

```python
pulocs['PULOC'] = [x.tolist() for x in pulocs['PULOC']]
droplocs['DropLoc'] = [x.tolist() for x in droplocs['DropLoc']]
```

```python
pulocsdict = dict(zip(pulocs['MissionID'], pulocs['PULOC']))
droplocsdict = dict(zip(droplocs['MissionID'], droplocs['DropLoc']))
```

```python
with open('../output/droplocsdict.yaml', 'w') as outfile:
        yaml.dump(droplocsdict,
                  outfile,
                  default_flow_style=False,
                  allow_unicode=True)
```

```python
with open('../output/pulocsdict.yaml', 'w') as outfile:
        yaml.dump(pulocsdict,
                  outfile,
                  default_flow_style=False,
                  allow_unicode=True)
```
