```python
import pandas as pd
import yaml
```

```python
with open('../input/passenger-dtypes.yaml', 'r') as yamlfile:
    dtypes = yaml.load(yamlfile)
```

```python
read_csv_opts = {'sep': '|',
                 'quotechar': '"',
                 'compression': 'gzip',
                 'encoding': 'utf-8',
                 'dtype': dtypes,
                 'parse_dates': ['MissionDate'],
                 'infer_datetime_format': True}
```

```python
fy11_passengers = '../input/ARTS_Passenger_Data_FY11-redacted.csv.gz'
fy12_passengers = '../input/ARTS_Passenger_Data_FY12-redacted.csv.gz'
fy13_passengers = '../input/ARTS_Passenger_Data_FY13-redacted.csv.gz'
fy14_passengers = '../input/ARTS_Passenger_Data_FY14-redacted.csv.gz'
fy15_passengers = '../input/ARTS_Passenger_Data_FY15-redacted.csv.gz'
fy16_passengers = '../input/ARTS_Passenger_Data_FY16-redacted.csv.gz'
fy17_passengers = '../input/ARTS_Passenger_Data_FY17-redacted.csv.gz'
fy18_passengers = '../input/ARTS_Passenger_Data_FY18-redacted.csv.gz'
fy19_passengers = '../input/ARTS_Passenger_Data_FY19_Year_To_Date-redacted.csv.gz'
```

```python
arts_passengers_fy11 = pd.read_csv(fy11_passengers, **read_csv_opts)
arts_passengers_fy12 = pd.read_csv(fy12_passengers, **read_csv_opts)
arts_passengers_fy13 = pd.read_csv(fy13_passengers, **read_csv_opts)
arts_passengers_fy14 = pd.read_csv(fy14_passengers, **read_csv_opts)
arts_passengers_fy15 = pd.read_csv(fy15_passengers, **read_csv_opts)
arts_passengers_fy16 = pd.read_csv(fy16_passengers, **read_csv_opts)
arts_passengers_fy17 = pd.read_csv(fy17_passengers, **read_csv_opts)
arts_passengers_fy18 = pd.read_csv(fy18_passengers, **read_csv_opts)
arts_passengers_fy19 = pd.read_csv(fy19_passengers, **read_csv_opts)
```

```python
passenger_files = [arts_passengers_fy11,
                 arts_passengers_fy12,
                 arts_passengers_fy13,
                 arts_passengers_fy14,
                 arts_passengers_fy15,
                 arts_passengers_fy16,
                 arts_passengers_fy17,
                 arts_passengers_fy18,
                 arts_passengers_fy19]
```

```python
df = pd.concat(passenger_files, sort=False)
```
